```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import style
from mpl_toolkits.mplot3d import Axes3D


df = pd.read_csv('/content/features.csv')


#exploring some data from the features file
print ('colunms:')
print (df.columns)


print ('firt_rows:')
print (df.head(10))


#taking a better look


# Selecting only the numeric columns
first_10_rows_numeric = df.select_dtypes(include=[np.number])

# Plotting the first 10 rows of all numeric columns as lines
for column in first_10_rows_numeric.columns:
    plt.plot(first_10_rows_numeric.index, first_10_rows_numeric[column], label=column)

plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Line Plot of First 10 Rows')
plt.legend()
plt.show()


#visualize the numerical data

numeric_df = df.iloc[:10, :].select_dtypes(include=[np.number])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#iterate over pairs of numerical columns
for i in range(len(numeric_df.columns)):
  for j in range(i+1, len(numeric_df.columns)):
    for k in range(j+1, len(numeric_df.columns)):

      #extract x,y,z data from data frame
      x = numeric_df.columns[i]
      y = numeric_df.columns[j]
      z = numeric_df.columns[k]

      #plot the data point
      ax.scatter(numeric_df[x], numeric_df[y], numeric_df[z])


#set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3d scatter plot of numerical data')

#show legend
ax.legend()
plt.show()


#doin the same with the sampleSub
df2 = pd.read_csv('/content/sampleSubmission.csv')


print ('colunms:')
print (df2.columns)


print ('firt_rows:')
print (df2.head(50))


#taking a better look


# Selecting only the numeric columns
first_10_rows_numeric2 = df2.select_dtypes(include=[np.number])

# Plotting the first 10 rows of all numeric columns as lines
for column in first_10_rows_numeric2.columns:
    plt.plot(first_10_rows_numeric2.index, first_10_rows_numeric2[column], label=column)


plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Line Plot of First 10 Rows')
plt.legend()
plt.show()

#lets open now and see the stores file

df3 = pd.read_csv('/content/stores.csv')

print ('columns:')
print (df3.columns)

print ('10_first_rows:')
print(df3.head(10))


#taking a better look


# Selecting only the numeric columns
first_10_rows_numeric2 = df2.select_dtypes(include=[np.number])

# Plotting the first 10 rows of all numeric columns as lines
for column in first_10_rows_numeric2.columns:
    plt.plot(first_10_rows_numeric2.index, first_10_rows_numeric2[column], label=column)




#extract x,y,z data from the subsect

plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Line Plot of First 10 Rows')
plt.legend()
plt.show()




#lets open and see some data from the test csv

df4 = pd.read_csv('/content/test.csv')

print ('columns:')
print (df4.columns)
print ('first_10_rows:')
print (df4.head(10))


#lets open now our last file the train
df5 = pd.read_csv('/content/train.csv')

print('columns:')
print (df5.columns)
print ('first_10_rows:')
print (df5.head(10))



```
